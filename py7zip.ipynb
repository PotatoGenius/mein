```python
# Import the py7zr library
import py7zr

# Specify the path to the 7z file
file_path = 'hashmob.net_2025-05-25.large.found.7z'

# Specify the directory to extract the contents to
extract_path = 'hashcat'

# Open the 7z file in read mode
with py7zr.SevenZipFile(file_path, mode='r') as archive:
    # Extract all contents to the specified directory
    archive.extractall(path=extract_path)

print(f'Successfully extracted {file_path} to {extract_path}')
```
